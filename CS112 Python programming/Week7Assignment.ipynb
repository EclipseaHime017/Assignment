# Bugzilla
Hideki is a new member of our DevOps group. The first task assigned to him is to do statistics on the accumulated bugs we left.  

He will be given a crap of bugs and their corresponding type, please help him count the number of each type.  

Input  
One line, the name of file you should read.

The file contains binary (pickled) representation of the bug list, in the form of:

[('cannot click on the login button', 'ui'), ('failed to login', 'auth'), ('cannot change the password', 'auth')]
The list contains several 2-ary tuples, the first element of each tuple is the bug's title, and the latter one is its type.

Output  
Multiple lines, you should count the number of bugs in each type, and output the result in CSV format. You need to first sort the result by the type name.

```python
import pickle
import os
os.chdir("E:\大一\python")
name = input()
category = {}
with open(name,'rb') as f:
    bugs = pickle.load(f)
    for item in bugs:
        x,y = item
        if y not in category:
            category[y] = 1
        elif y in category:
            category[y] += 1
    keys = category.keys()
    values = category.values()
    zipped = zip(keys,values)
    nested = list(zipped)
    for item in nested:
        print(f"{item[0]},{item[1]}")



    
```

# Asterisk Triangle
Description
Given an integer N,   
N is an odd number, and output an equilateral triangle consisting of ∗∗.

The first row has 1 ∗∗, the second row has 3 ∗∗, 

the third row has 5 ∗∗, and so on. The last row has a total of N ∗∗.

Note: The ' ' to the right of each line 
∗
∗ is necessary.

```python
filename = input()
with open(filename,"r")as fh:
    amount = int(fh.read())
    layer = int((amount+1)/2)
    for a in range(1,layer+1):
        forma = " "*(layer-a)+"*"*int(a*2-1)+" "*(layer-a)
        print(f'{forma}')
```

 # Class Rank
The mid-term exam is over, and the teacher wants to collect information about the exam results and rank the students' total scores. Help the teacher.  

Input  
The input is the path name of the file.  

filename = input()

Output  
The output is a list containing the ranking information of students(rank students' total scores from high to low).  
If the scores of two students are the same, sort by name.

```python
import csv
filename = input()
student = []
total = 0
newlist = []
lis = []
result = []
with open(filename,"r")as fh:
    csv_student = csv.reader(fh)
    for elmnt in csv_student:
        student.append(elmnt)
sta = student[1:]
for grade in sta:
    for n in range(1,5):
        total += int(grade[n])
    newlist = [total,grade[0]]
    lis.append(newlist)
    newlist = []
    total = 0
lis.sort(reverse=True)
for person in lis:
    result.append(person[1])
print(result)
        
```
